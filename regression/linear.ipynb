# Linear regression

### Using numpy's polyfit and polyval for OLS regression on simple monomials

```python
import os
import random
import numpy as np
from pylab import mpl, plt
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
os.environ['PYTHONHASHSEED'] = '0'
```

```python
x = np.linspace(0, 10)
x
```

```python
def set_seeds(seed=100):
    random.seed(seed)
    np.random.seed(seed)
set_seeds()
```

```python
y = x + np.random.standard_normal(len(x))
fit_result = np.polyfit(x, y, deg=1)
fit_result
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', label='data')
plt.plot(x, np.polyval(fit_result, x), 'r', lw=2, label='linear')
plt.legend(loc=0);
```

### Extrapolate

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', label='data')
xn = np.linspace(0, 20)
plt.plot(xn, np.polyval(fit_result, xn), 'r', lw=2, label='linear')
plt.legend(loc=0);
```

### Basics for time series

```python
# split the time series so that each part has 3 inputs and 1 output
x = np.arange(12)
lags = 3
m = np.zeros((lags + 1, len(x) - lags))
for i in range(lags + 1):
    m[i] = x[i:len(x) - lags + i]
m
```

```python
result = np.linalg.lstsq(m[:lags].T, m[lags], rcond=None)[0]
result
```

```python
np.dot(m[:lags].T, result)
```

### Using numpy's linalg.lstsq for general least-square regression

```python
import pandas as pd
raw = pd.read_csv('../data/pyalgo_eikon_eod_data.csv',
                 inlinear_model.LogisticRegressiondex_col=0, parse_dates=True).dropna()
raw.head()
```

```python
symbol = 'EUR='
data = pd.DataFrame(raw[symbol])
data.rename(columns={symbol:'price'}, inplace=True)
```

```python
lags = 5
cols = []
for lag in range(1, lags + 1):
    col = f'lag_{lag}' # the same as: col = 'lag_%s' % lag
    data[col] = data['price'].shift(lag)
    cols.append(col)
data.dropna(inplace=True)
data.head()
```

```python
# use the lag 5 to predict the spot 
fit_result = np.linalg.lstsq(data[cols], data['price'],
                             rcond=None)[0]
fit_result
```

```python
data['prediction'] = np.dot(data[cols], fit_result)
data[['price', 'prediction']].plot(figsize=(10, 6));
```

```python
data[['price', 'prediction']].loc['2019-10-1':].plot(figsize=(10, 6));
```

### Predicting future returns
#### Apply linear regression to log returns

```python
data['return'] = np.log(data['price'] / data['price'].shift(1))
data.dropna(inplace=True)
cols = []
for lag in range(1, lags + 1):
    col = f'lag_{lag}'
    data[col] = data['return'].shift(lag)
    cols.append(col)
data.dropna(inplace=True)
data[cols].head()
```

```python
fit_result = np.linalg.lstsq(data[cols], data['return'], rcond=None)[0]
fit_result
```

```python
data['prediction'] = np.dot(data[cols], fit_result)
data[['return', 'prediction']].iloc[lags:].plot(figsize=(10, 6));
```

```python
correct_prediction_count = np.sign(data['return'] * data['prediction']).value_counts()
```

```python
correct_prediction_count
```

### Predicting future market direction
#### Apply regression on the **sign** of the log returns

```python
fit_result = np.linalg.lstsq(data[cols], np.sign(data['return']), rcond=None)[0]
fit_result
```

```python
data['prediction'] = np.sign(np.dot(data[cols], fit_result))
data['prediction'].value_counts()
```

```python
hits = np.sign(data['return'] * data['prediction']).value_counts()
hits
```

```python
hits_ratio = hits.values[0] / sum(hits)
hits_ratio
```
